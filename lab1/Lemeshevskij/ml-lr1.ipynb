```python
import pandas as pd
from matplotlib import pyplot as pl
%matplotlib inline
df = pd.read_csv('district_wise_rainfall_normal.csv')
df.head()
```

## Задание 1
#### Постройте графики количества осадков по годам в разных штатах. Сделайте выводы по построенным графикам.

```python
pl.bar(df['DISTRICT'].index, height=df['ANNUAL'])
pl.xlabel(df['DISTRICT'], rotation=90)
pl.show()
```

## Задание 2
#### Постройте графики количества осадков по месяцам по всем годам кумулятивно. Сделайте выводы по построенным графикам.

```python
mr = pd.read_csv('rainfall_in_india_1901-2015.csv')
mr.head()
```

```python
tmp_mr = mr.groupby(['YEAR']).agg('sum').sum()
pl.bar(range(12), height=tmp_mr.values[:12], tick_label=tmp_mr.keys()[:12])
pl.show()
```
